```python
## Question - 4
teams = {
    'Kolkata Knight Riders': {'Players': [], 'Purse': 32.70, 'PurseUsed': 0},
    'Sunrisers Hyderabad': {'Players': [], 'Purse': 34.00, 'PurseUsed': 0},
    'Chennai Super Kings': {'Players': [], 'Purse': 31.40, 'PurseUsed': 0},
    'Gujarat Titans': {'Players': [], 'Purse': 38.15, 'PurseUsed': 0},
    'Punjab Kings': {'Players': [], 'Purse': 29.10, 'PurseUsed': 0}
}

def add_player(team_name, player_name, selling_price):
    if team_name in teams and teams[team_name]['Purse'] >= selling_price:
        teams[team_name]['Players'].append({'PlayerName': player_name,
                                            'SellingPrice': selling_price})
        teams[team_name]['Purse'] -= selling_price
        teams[team_name]['PurseUsed'] += selling_price
        print(f"{player_name} added to {team_name} for {selling_price:0.2f} Cr.")
    else:
        print(f"Error: Insufficient purse balance for {team_name} or invalid team name.")

def sort_teams_by_selling_value():
    sorted_teams = sorted(teams.items(), key=lambda x: x[1]['PurseUsed'],
                          reverse=True)
    return sorted_teams

def display_teams():
    for team_name, details in teams.items():
        print(f"{team_name}: Purse Remaining - {details['Purse']:0.2f} Cr,Purse Used - {details['PurseUsed']} Cr")
        print("Players:")
        for player in details['Players']:
            print(f"  {player['PlayerName']}: {player['SellingPrice']} Cr")
        print("\n")

# Example Usage:
add_player('Kolkata Knight Riders', 'Mitchell Starc', 24.75)
add_player('Kolkata Knight Riders', 'Mujeeb Ur Rahman', 2.00)
add_player('Chennai Super Kings', 'Sameer Rizvi', 8.40)
add_player('Chennai Super Kings', 'Daryl Mitchell', 14.00)
add_player('Gujarat Titans', 'Spencer Johnson', 10.00)
add_player('Sunrisers Hyderabad', 'Travis Head', 6.80)
add_player('Punjab Kings', 'Harshal Patel', 11.75)
add_player('Sunrisers Hyderabad', 'Pat Cummins', 20.50)
add_player('Gujarat Titans', 'Shahrukh Khan', 7.40)
add_player('Punjab Kings', 'Rilee Rossouw', 8.00)


sorted_teams = sort_teams_by_selling_value()

print("\nTeams after sorting by selling value:")
for team_name, details in sorted_teams:
    print(f"{team_name}: Purse Remaining - {details['Purse']:0.2f} Cr, Purse Used - {details['PurseUsed']:0.2f} Cr")
    print("Players:")
    for player in details['Players']:
        print(f"  {player['PlayerName']}: {player['SellingPrice']} Cr")
    print("\n")
```
